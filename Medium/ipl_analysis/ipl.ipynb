```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv("matches.csv")
matches.head()
```

```python
matches.columns
```

```python
# out of the toss wins what is the probabilty of batting and bowling
bat = matches.loc[matches["toss_decision"]=="bat"].count()[0]
print(bat)

field = matches.loc[matches["toss_decision"]=="field"].count()[0]
print(field)

total = bat + field
print(total)

# calculating the percentage of batting and fielding
batting = bat/total * 100
print(batting)

fielding = field/total*100
print(fielding)
```

```python
#plot the pie chart for above analysis

plt.figure(figsize=(4,2))

plt.title("Probabiltiy of toss winners to choose batting or fielding ")

labels = ["batting","fielding"]
plt.pie([batting,fielding],labels=labels,autopct="% .2f %%")

plt.show()
```

```python
# who has hold the most no of player of match in each season
 
#group by season and pom and count occurences
pom_counts=matches.groupby(["season","player_of_match"]).size().reset_index(name="count")
pom_counts

# get the top players each season 
top_player =pom_counts.loc[pom_counts.groupby("season")['count'].idxmax()] #idmax is used to get the highest number 
top_player
```

```python
# plot the pom
plt.figure(figsize=(12,5))
plt.title("Top player of match of each ipl season",fontdict=
          {"fontsize":10, "fontweight":"bold"})
sns.set_style("darkgrid")
plt.xlabel("Season")
plt.ylabel("No of awards")
sns.barplot(x="season",y="count",hue="player_of_match",data=top_player,dodge=False,edgecolor = "black")

plt.xticks(rotation=45)
plt.show()
```

```python
# In which inning do players score the most ?
innings =ipl.groupby("inning")["total_runs"].mean()
print(innings)

plt.figure(figsize=(10,5))

plt.title("Distribution of runs scored in each innings",fontdict={"fontsize":16,"color":"red"})
bars =sns.barplot(x="inning",y="total_runs",data=ipl.groupby("inning")["total_runs"].sum().reset_index())

plt.xlabel("innings" ,fontdict={"fontsize" : 15})
plt.ylabel("Runs Scored" ,fontdict={"fontsize" : 15})

plt.show()
```

```python
#In which stadium has Mumbai Indians won its majority games
mi_wins= matches[matches["winner"]=="Mumbai Indians"]
mi_wins

venue_count = mi_wins["venue"].value_counts().head()
print(venue_count)

most_wins_venue = venue_count.idxmax()
most_wins_venue
```

```python
#plotting top 5 stadiums Mumbai Indians have won its majority matches 
plt.figure(figsize=(8,4))
plt.title("Top 5 stadiums MI has won its matches",fontdict={"fontsize":15,
                                                            "color":"red"})
colors = ["lightblue","orange","grey","yellow","pink"]
plt.pie(venue_count,labels=venue_count.index,colors=colors,wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
#In which stadium has Kings XI Punjab won its majority games
pj_wins= matches[matches["winner"]=="Kings XI Punjab"]
pj_wins

venue_count = pj_wins["venue"].value_counts().head()
print(venue_count)

most_wins_venue = venue_count.idxmax()
most_wins_venue
```

```python
#total wickets taken in each match 
wickets=deliveries[deliveries["player_dismissed"].notna()]

wickets_taken=wickets.groupby("match_id").size().reset_index(name="total_wickets")

wickets_taken
```

```python
plt.figure(figsize=(8,4))
match_ids = deliveries["match_id"].head()
plt.title("Kde plot for Total wickets Taken")
sns.kdeplot(wickets_taken["total_wickets"],shade=True)
plt.show()
```

```python
#Most succesfull Teams
matches["winner"].value_counts().head()
```

```python
#wickets vs Season vs match 
wickets_with_season=pd.merge(wickets_taken,matches[["id","season"]],left_on="match_id",right_on="id")
wickets_with_season
```

```python
pivot = wickets_with_season.pivot_table(index="season",columns="match_id",values="total_wickets")
sns.heatmap(pivot)
```
