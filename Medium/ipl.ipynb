```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv("matches.csv")
matches.head()
```

```python
deliveries = pd.read_csv("deliveries.csv")
deliveries.head()
```

```python
# getting the total runs scored by virat Kohli in the ipl t20 format fromm 2008 - 2024
print(deliveries[deliveries["batter"]=="V Kohli"]["batsman_runs"].sum())

# getting the total runs scored by Rohit Sharma in the ipl t20 format fromm 2008 - 2024
print(deliveries[deliveries["batter"]=="RG Sharma"]["batsman_runs"].sum())

# getting the total runs scored by Rohit Sharma in the ipl t20 format fromm 2008 - 2024
print(deliveries[deliveries["batter"]=="DJ Bravo"]["batsman_runs"].count())


```

```python
matches.columns
```

```python
# out of the toss wins what is the probabilty of batting and bowling
bat = matches.loc[matches["toss_decision"]=="bat"].count()[0]
print(bat)

field = matches.loc[matches["toss_decision"]=="field"].count()[0]
print(field)

total = bat + field
print(total)

# calculating the percentage of batting and fielding
batting = bat/total * 100
print(batting)

fielding = field/total*100
print(fielding)
```

```python
#plot the pie chart for above analysis

plt.figure(figsize=(4,2))

plt.title("Probabiltiy of toss winners to choose batting or fielding ")

labels = ["batting","fielding"]
plt.pie([batting,fielding],labels=labels,autopct="% .2f %%")

plt.show()
```

```python
# who has hold the most no of player of match in each season
 
#group by season and pom and count occurences
pom_counts=matches.groupby(["season","player_of_match"]).size().reset_index(name="count")
pom_counts

# get the top players each season 
top_player =pom_counts.loc[pom_counts.groupby("season")['count'].idxmax()] #idmax is used to get the highest number 
top_player
```

```python
# plot the pom
plt.figure(figsize=(12,5))
plt.title("Top player of match of each ipl season",fontdict=
          {"fontsize":10, "fontweight":"bold"})
sns.set_style("darkgrid")
plt.xlabel("Season")
plt.ylabel("No of awards")
sns.barplot(x="season",y="count",hue="player_of_match",data=top_player,dodge=False,edgecolor = "black")

plt.xticks(rotation=45)
plt.show()
```
