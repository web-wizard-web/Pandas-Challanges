# 3. COVID-19 Daily Cases
**Objective: Track and visualize daily confirmed, death, and recovered cases globally.**

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid = pd.read_csv("country_wise_latest.csv")
covid.head()
```

```python
covid[covid["Country/Region"]== "Afghanistan"]
```

```python
# plot the death and recovered cases
plt.figure(figsize=(10, 6))
# to get only few countries
countries = ["Afghanistan", "Albania", "Algeria", "Andorra", "Angola", "Antigua and Barbuda", "Argentina", "Armenia", "Australia", "Austria"]
# filter the dataframe for the selected countries
filtered_covid = covid[covid["Country/Region"].isin(countries)]
sns.barplot(x="Country/Region", y="Deaths" ,data=filtered_covid ,color="red")

# title
plt.title("Death Cases in Selected Countries")
plt.xticks(rotation=45)

plt.show()
```

```python
# Filter the countries having death greter than 10000
death_1000 = covid[covid["Deaths"]>10000]
death_1000
```

```python
# Death cases in India


india_death = covid[covid["Country/Region"]=="India"]["Deaths"].values[0]
```

```python
#Recovery rate of top 20 countries
covid["Recovery Percentage"]= covid["Recovered"]/covid["Confirmed"].astype(int)

# Sorting the dataframe by recovery percentage
top_countries = covid.sort_values("Recovery Percentage", ascending=False).head(20)

# Plotting the recovery percentage
plt.figure(figsize=(10,5))
sns.barplot(x="Country/Region", y="Recovery Percentage", data=top_countries )
plt.xticks(rotation=45)

# tight_layout() to adjust the layout
plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Plotting the death percentage of top 20 countries 
covid["Death percentage"] = covid["Deaths"]/covid["Confirmed"].astype(float)

top_countries_death = covid.sort_values("Deaths", ascending= False).head(20)

# Plotting the death percentage

plt.figure(figsize=(10,5))

plt.title("Death percentage of 20 countries")

sns.barplot(x="Country/Region", y="Death percentage", data=top_countries_death , color="red")

plt.xticks(rotation=45)

plt.show()

```

```python
# Get the new data for continet wise analysis
covid_continent = pd.read_csv("covid_worldometer_data.csv")
covid_continent.head()

```

```python
# Analyzing the data for continent Asia
asia_data = covid_continent[covid_continent["Continent"]=="Asia"]
asia_data

# create csv for asia data
asia_data.to_csv("asia_covid_data.csv", index=False)
```

```python
# Death cases in Asia's top 10 populated countries 
populated = asia_data.sort_values("Population" , ascending = False).head(10)

plt.figure(figsize=(10,5))
plt.title("Deaths cases in Asias 10 populated countries")
sns.lineplot(x="Country/Region" , y="TotalDeaths", data=populated, color="red", marker="o")

plt.xlabel("Countries")
plt.ylabel("Total Deaths")

plt.show()
```

```python
# Analyzing the data for continent Asia
europe_data = covid_continent[covid_continent["Continent"]=="Europe"]
europe_data

# create csv for asia data
europe_data.to_csv("europe_covid_data.csv", index=False)
```

```python
# Death cases in Asia's top 10 populated countries 
populated_euro = europe_data.sort_values("Population" , ascending = False).head(10)

plt.figure(figsize=(10,5))
plt.title("Deaths cases in Asias 10 populated countries")
sns.lineplot(x="Country/Region" , y="TotalDeaths", data=populated_euro, color="blue", marker="*")

plt.xlabel("Countries")
plt.ylabel("Total Deaths")

plt.show()
```

**Continent Level Aggregation**

```python
# Plotting it using groupby
continent_group = covid_continent.groupby("Continent").sum()
continent_group
```

```python
# Plotting deaths cases by one million population of each continent
sns.barplot(data=continent_group, x=continent_group.index, y="Tot Cases/1M pop", color="green")

plt.title("Total Deaths by Continent")
plt.xticks(rotation=50)

plt.show()
```
