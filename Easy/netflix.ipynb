**Objective: Explore the distribution of Netflix titles by genre, country, and release year.**

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv('netflix_titles.csv')
netflix.head()
```

```python
# Get the columns of series produces in India
india = netflix[netflix["country"]=="India"][["title","director"]]
india
```

```python
# Get the tv shows released  in the year 2020
netflix[(netflix["listed_in"]== "TV Shows" ) & (netflix["release_year"]==2020)]
```

```python
# using pivot to reshape data
pivot = netflix.pivot_table(columns="rating", index="country" ,values="title", aggfunc="count")
pivot
```

```python
#Number of titles added per release year 

#aggfunc is used to aggregate data when multiple rows fall in same rows+column group 
pivot2 = netflix.pivot_table(index="release_year", columns="rating",values="title",aggfunc="count") 
pivot2
```

```python
#getting all the different rating
netflix["rating"].unique()
```

```python
# Clleaning the rating column (removing the unnecessary timing)
valid_ratings = ['PG-13', 'TV-MA', 'PG', 'TV-14', 'TV-PG', 'TV-Y', 'TV-Y7', 'R',
       'TV-G', 'G', 'NC-17' 'NR', 'TV-Y7-FV', 'UR']

netflix["cleaned_ratings"]= netflix["rating"].where(netflix["rating"].isin(valid_ratings),np.nan) 
```

```python
# Add a column to explain what the rating means 
conditions = [
    netflix["rating"]== "TV-MA",
    netflix["rating"]== "TV-14",
    netflix["rating"]== "TV-PG",
    netflix["rating"]== "PG-13",
    netflix["rating"]== "PG",
    netflix["rating"]== "TV-Y",
    netflix["rating"]== "TV-Y7",
    netflix["rating"]== "R",
    netflix["rating"]== "TV-G",
    netflix["rating"]== "G",
    netflix["rating"]== "NC-17",
    netflix["rating"]== "NR",
    netflix["rating"]== "TV-YZ-FV",
    netflix["rating"]== "UR",
]

choices = ["Matured Audience ONly - Adults only",
           "Parents Strongly Cautonied -14+",
           "Parental Guidance Suggested",
           "Parents Strongly Cautonied -13+",
           "Parental Guidance",
           "Suitable for all Children ",
           "Age 7 and Up",
           "Restricted -17 + only with adult",
           "Television general auidiance - Suitable for all",
           "General Audiance - All ages",
           "Adults Only -17+",
           "Not Rated",
           "age above seven but fantasy violence",
           "Unrated"
           ]

netflix["rating explained"] = np.select(conditions,choices,default="No description available")
netflix.head()
```
